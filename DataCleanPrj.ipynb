**Visualizations starts here**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.Sales over time

df['Order Date'] = pd.to_datetime(df['Order Date'],format="%d-%m-%Y")
monthlySales = df.resample('m',on='Order Date')['Sales'].sum()
plt.figure(figsize=(12,6))
plt.plot(monthlySales.index,monthlySales)
plt.title("Total Sales per Month")
plt.xlabel("Months")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
# 2. Profit By Category
categoryProfit = df.groupby('Category')['Profit'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(data=categoryProfit,x='Category',y='Profit',palette='coolwarm')
plt.title("Profit by Category ")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.grid()
plt.show()
```

```python
# 3 Sales Distribution By region
regionalSales = df.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(data=regionalSales,x='Region',y='Sales',palette='viridis')
plt.title("Sales by Region ")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
# 4 Profits Distribution By region
regionalProfits = df.groupby('Region')['Profit'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(data=regionalProfits,x='Region',y='Profit',palette='coolwarm')
plt.title("Profit by Region ")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.grid()
plt.show()
```

```python
# 5. Coorealtion Heat map
numerical_features = df[['Sales','Profit','Quantity']]
sns.heatmap(numerical_features.corr(),annot=True)
```

```python
#6.  Sales and Profits by Segment
aggregated_data = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a bar chart for Sales
plt.bar(aggregated_data['Segment'], aggregated_data['Sales'], color='blue', alpha=0.6, label='Sales')

# Create a bar chart for Profits
plt.bar(aggregated_data['Segment'], aggregated_data['Profit'], color='green', alpha=0.6, label='Profit')

# Adding titles and labels
plt.title('Sales and Profits by Segment')
plt.xlabel('Segment')
plt.ylabel('Amount ($)')
plt.legend()

# Display the plot
plt.show()
```

```python
#7. Product Category Sales Distribution
product_category_sales = df.groupby('Category')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=product_category_sales, x='Category', y='Sales', palette='viridis')
plt.title('Product Category Sales Distribution')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
#8. Top 10 products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='Product Name', y='Sales', palette='viridis')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
#9 . Sales vs Quantity Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Quantity', hue='Segment', palette='viridis')
plt.title('Sales vs Quantity')
plt.xlabel('Sales')
plt.ylabel('Quantity')
plt.show()
```
