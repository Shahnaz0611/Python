Feature Engineering

```python
df['storage_encoded']=df['HDD']+df['SSD']+df['Flash Storage']
# Creates a new column 'storage_encoded' by summing the values of 'HDD', 'SSD', and 'Flash Storage'

df.drop(['HDD','SSD','Flash Storage'],axis=1,inplace=True)
# Drops the columns 'HDD', 'SSD', and 'Flash Storage' from the DataFrame inplace

df.drop(['Hybrid'],axis=1,inplace=True)
# Drops the column 'Hybrid' from the DataFrame inplace
```

```python
df.columns
```

```python
# This line creates a new DataFrame `touch_embeded` by filtering rows from `df` where the 'Touchscreen' column has the value 0.
touch_embeded = df[df['Touchscreen'] == 0]
```

```python
# This line drops the rows from the DataFrame `df` where the 'Touchscreen' column has the value 0.
df.drop(df[df['Touchscreen'] == 0].index, inplace=True)
```

```python
# This line drops the 'Touchscreen' column from the DataFrame `df`.
df.drop(['Touchscreen'], axis=1, inplace=True)
```
